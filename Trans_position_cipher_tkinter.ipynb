```python
from tkinter import *
from math import ceil

window = Tk()
window.title("Encrption and Decription")
window.geometry("500x700")
# window.resizable(False, False)

heading_one = Label(window,text='ENCRIPTION',font=('TimesNewRoman',17,'bold','underline'), fg="Red")
heading_one.place(x=200,y=100)

encrpt_code = Label(window, text= 'Enter Encrption Code').place(x=125, y=150)
encrpt_code_entry = Entry(window)
encrpt_code_entry.place(x=250, y=150)

encrpt_message = Label(window, text= 'Enter Message').place(x=125, y=200)
encrpt_mess_entry = Entry(window)
encrpt_mess_entry.place(x=250, y=200)

# Devrption

heading_two = Label(window,text='DECRIPTION',font=('TimesNewRoman',17,'bold','underline'), fg="Red")
heading_two.place(x=200,y=400)

decrpt_code = Label(window, text= 'Enter Decrption Code').place(x=125, y=450)
decrpt_code_entry = Entry(window)
decrpt_code_entry.place(x=250, y=450)

decrpt_message = Label(window, text= 'Enter Message').place(x=125, y=500)
decrpt_mess_entry = Entry(window)
decrpt_mess_entry.place(x=250, y=500)

def encrypt_message(message, key):
    message_split = list(message.lower())
    message_length = len(message_split)
    message_encrypted = ''
    message_ceil = ceil(message_length/int(key))
    for j in range(int(key)):
        for i in range(message_ceil):
            index = j + i * int(key)
            if index < message_length:
                message_encrypted += message_split[index]
    output = message_encrypted
    encrpted_message= Label(window, text = f'Encrpted Message :{output}|').place(x=120, y=300)
    
def func():
    var = encrypt_message(encrpt_mess_entry.get(), key = encrpt_code_entry.get())
    return var


def decrypt_message(message, key2):
    message_split = list(message.lower())
    message_length = len(message_split)
    message_ceil = ceil(message_length/int(key2))
    num_empty_cells = int(key2)*message_ceil - message_length
    message_grid = [['' for _ in range(message_ceil)] for _ in range(int(key2))]
    message_decrypted = ''
    iterator = iter(message_split)
    for i in range(int(key2)):
        if i < int(key2) - num_empty_cells:
            columns = message_ceil
        else:
            columns = message_ceil - 1
        for j in range(columns):
            message_grid[i][j] = next(iterator, None)
    for j in range(message_ceil):
        for i in range(int(key2)):
            message_decrypted += message_grid[i][j]
    output2 = message_decrypted
    decrpted_message= Label(window, text = f'Decrpted Message :  {output2}').place(x=120, y=600)

def func2():
    var = decrypt_message(decrpt_mess_entry.get(), key2 = decrpt_code_entry.get())
    return var

submit1 = Button(window, text="ENCRPT", command = func).place(x=150, y=250)
submit2 = Button(window, text="DECRPT", command = func2).place(x=150, y=550)

window.mainloop()
```
